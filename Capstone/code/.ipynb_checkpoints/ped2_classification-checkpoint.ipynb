## Part 2 (Image Classification)  

The second notebook addresses the first goal set in the problem statement, namely to develop an image classification model with the ability to distinguish between persons and non-persons. This will be done via training the dataset on customized self-created CNN models followed by pre-trained CNN models.

### 2.1 Image Classification via Convolutional Neural Network

In preparation for model training, we first augment all images using Keras' ImageDataGenerator Function. Augmentation involves a mixture of slight rotations, flipping images, small random crops etc., intended to add slight "disturbances" without influencing the main object(s) so that the convolutional neural network (CNN) is more robust to real-world variations. Further, this pre-processing step is required when passing images into pre-trained CNN models. 


```python
# Load in resorted images with correct labels using Keras' ImageDataGenerator Function
## Normalize images by dividing each pixel by 255 to scale value between 0 and 1
imagegen = ImageDataGenerator(rescale = 1. / 255) 
train_data = imagegen.flow_from_directory("data/sorted_kaggle/train",  
                                          target_size = (224, 224), 
                                          class_mode = "binary", 
                                          shuffle = True, 
                                          batch_size = 128)

val_data = imagegen.flow_from_directory("data/sorted_kaggle/val", 
                                          target_size = (224, 224), 
                                          class_mode = "binary", 
                                          shuffle = True, 
                                          batch_size = 128)

test_data = imagegen.flow_from_directory("data/sorted_kaggle/test", 
                                          target_size = (224, 224), 
                                          class_mode = "binary", 
                                          shuffle = True, 
                                          batch_size = 128)
```

```python
# Checking binary class
train_data.class_indices
 
```

```python
# Create X and y sets in case required for later use
X_train, y_train = next(train_data)
X_val, y_val = next(val_data)
X_test, y_test = next(test_data)
```

```python
# Create custom function to generate f1-score and confusion matrix
def metrices(model): # Argument = model type
    preds = np.round(model.predict(X_test), 0) # Apply model to predict test set
    print("F1 Score: " + str(f1_score(y_test, preds))) # Generate F1 score
    sns.heatmap(confusion_matrix(y_test, preds), cmap = "Blues", annot = True) # Generate confusion matrix (tn, fp, fn, tp)
    
```

<div id="base"></div>

##### 2.1.1 Base

First, we commence training with a basic CNN model comprising a single convolutional/pooling stack and single hidden layer: 

```python
model = Sequential()
# Tested with filters = 4, 8 and 16 
model.add(Conv2D(filters = 8, kernel_size = (3, 3), activation = 'relu', input_shape = (224, 224, 3))) 
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())
model.add(Dense(128, activation = 'relu'))
# Model perforamnce increase when applied due to regularization and reduced generalization errors
model.add(BatchNormalization()) 
model.add(Dense(1, activation = 'sigmoid'))
# Tested with lr = 0.01, 0.001 and 0.0001
optimizer1 = Adam(learning_rate = 0.001) 
model.compile(optimizer = optimizer1, loss = 'binary_crossentropy', metrics = ['accuracy', 'Recall'])
```

```python
history = model.fit(train_data, batch_size = 256, epochs = 50, validation_data = val_data)
```

```python
model.summary()
```

```python
# Check train loss and val loss over epochs
train_loss = history.history['loss']
test_loss = history.history['val_loss']
epoch_labels = history.epoch

plt.figure(figsize = (12, 8))

plt.plot(train_loss, label = 'Training Loss', color = '#185fad')
plt.plot(test_loss, label = 'Val Loss', color = 'orange')

plt.title('Training and Val Loss by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Binary Crossentropy', fontsize = 18)
plt.xticks(epoch_labels, epoch_labels)    

plt.legend(fontsize = 18);
```

```python
# Check out our train and val accuracy over epochs
train_loss = history.history['accuracy']
test_loss = history.history['val_accuracy']
epoch_labels = history.epoch

plt.figure(figsize = (12, 8))

plt.plot(train_loss, label = 'Training Accuracy', color = '#185fad')
plt.plot(test_loss, label = 'Val Accuracy', color = 'orange')

plt.title('Training and Val Accuracy by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Accuracy', fontsize = 18)
plt.xticks(epoch_labels, epoch_labels)    

plt.legend(fontsize = 18);
```

```python
# Score on test set and generate metrics
score = model.evaluate(test_data, verbose = 1) 
metrices(model)
```

## 2.2 Image Classification via Transfer Learning

Transfer learning is a technique where a model trained on a given ML task is repurposed and reused as a starting/intermediate point for a different but related job ([*source*](https://www.learndatasci.com/tutorials/hands-on-transfer-learning-keras/)); the "transfer" of knowledge from one model to another allows rapid progress and improved performance. It is typically applied for deep learning given the large amount of resources required to comprehensively train neural networks, which includes but is not limited to the huge and challenging datasets involved ([*source*](https://machinelearningmastery.com/transfer-learning-for-deep-learning/)). For the purposes of this study, two pre-trained models will be analyzed: VGG16 and ResNet50. 


<div id="vgg16"></div>

##### 2.2.1 VGG16

VGG is the abbreviation for Visual Geometry Group, a group of researchers from the University of Oxford who developed the renowned neural network model encompassing 16 layers, hence the term VGG16 ([*source*](https://medium.com/@mygreatlearning/what-is-vgg16-introduction-to-vgg16-f2d63849f615)). VGG16 is trained on ImageNet, a dataset of over 14 million images belonging to 1,000 classes. In terms of architecture, there are 3 stacks of convolutional/pooling each their respective parameters followed by 5 fully connected (i.e. hidden) layers. 

<img src = "../image/vgg16.png" alt = "vgg16_archi"/>

*Fig. 6 VGG16 Architecture* ([*source*](https://miro.medium.com/max/940/1*3-TqqkRQ4rWLOMX-gvkYwA.png))


```python
# Load VGG16 as base model
vgg16_model = VGG16(include_top = False, weights = 'imagenet', input_shape = (224, 224, 3))
vgg16_model.summary()
```

```python
# Freeze all pre-trained layers within VGG16 model
for layer in vgg16_model.layers:
    layer.trainable = False
    
```

```python
# Generate top layers based on dataset in question
top_model = vgg16_model.output
top_model = GlobalAveragePooling2D()(top_model)
top_model = Flatten()(top_model)
top_model = Dense(128, activation='relu')(top_model)
top_model = BatchNormalization()(top_model)
output_layer = Dense(1, activation='sigmoid')(top_model)
    
```

```python
# Merge VGG16 base model with self-created top layers
model7 = Model(inputs = vgg16_model.input, outputs = output_layer)
```

```python
# Compile model
optimizer1 = Adam(learning_rate = 0.001) 
model7.compile(optimizer = optimizer1, loss = 'binary_crossentropy', metrics = ['accuracy', 'Recall'])
```

```python
history7 = model7.fit(train_data, batch_size = 256, epochs = 50, validation_data = val_data)
```

```python
model7.summary()
```

```python
# Check train loss and val loss over epochs
train_loss = history7.history['loss']
test_loss = history7.history['val_loss']
epoch_labels = history7.epoch

plt.figure(figsize = (12, 8))

plt.plot(train_loss, label = 'Training Loss', color = '#185fad')
plt.plot(test_loss, label = 'Val Loss', color = 'orange')

plt.title('Training and Val Loss by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Binary Crossentropy', fontsize = 18)
plt.xticks(epoch_labels, epoch_labels)    

plt.legend(fontsize = 18);
```

```python
# Check train loss and val accuracy over epochs
train_loss = history7.history['accuracy']
test_loss = history7.history['val_accuracy']
epoch_labels = history7.epoch

plt.figure(figsize = (12, 8))

plt.plot(train_loss, label = 'Accuracy', color = '#185fad')
plt.plot(test_loss, label = 'Val Accuracy', color = 'orange')

plt.title('Training and Val Accuracy by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Accuracy', fontsize = 18)
plt.xticks(epoch_labels, epoch_labels)   

plt.legend(fontsize = 18);
```

```python
# Score on test set and generate metrics
score = model7.evaluate(test_data, verbose = 1)
metrices(model7)
```

##### 2.2.2 ResNet50

ResNet is short for Residual Network and as implied, contains 50 layers internally. Likewise, ResNet50 is trained on the ImageNet dataset. Traditionally, extremely deep neural networks were deemed unfeasible due to the problem of vanishing gradients: "as the gradient is back-propagated to earlier layers, repeated multiplication may make the gradient extremely small. As a result, as the network goes deeper, its performance gets saturated or even starts degrading rapidly" ([*source*](https://towardsdatascience.com/understanding-and-coding-a-resnet-in-keras-446d7ff84d33)). As such, the main strength of ResNet50 is its 'skip connection' ability, whereby selected layers in the neural network are skipped and the output of a given layer is directly fed as the input to the following layer, in turn allowing gradient degradation to be mitigated ([*source*](https://www.analyticsvidhya.com/blog/2021/08/all-you-need-to-know-about-skip-connections/)). The ResNet50 architecture contains 48 convolutional layers along with 1 Max Pooling layer and 1 Average Pooling layer ([*source*](https://iq.opengenus.org/resnet50-architecture/)). 


<img src = "../image/rn50.png" alt = "rn50_archi"/>

*Fig. 7 ResNet50 Architecture* ([*source*](https://cdn-5f733ed3c1ac190fbc56ef88.closte.com/wp-content/uploads/2019/07/ResNet50_architecture-1.png))


```python
# Load ResNet50 as base model
rn50_model = ResNet50(include_top = False, weights = 'imagenet', input_shape = (224, 224, 3))
rn50_model.summary()
```

```python
# Freeze all pre-trained layers within ResNet50 model
for layer in rn50_model.layers:
    layer.trainable = False
    
```

```python
# Generate top layers based on dataset in question
top_model = rn50_model.output
top_model = GlobalAveragePooling2D()(top_model)
top_model = Flatten()(top_model)
top_model = Dense(128, activation = 'relu')(top_model)
top_model = BatchNormalization()(top_model)
output_layer = Dense(1, activation = 'sigmoid')(top_model)
    
```

```python
# Merge ResNet50 base model with self-created top layers
model8 = Model(inputs = rn50_model.input, outputs = output_layer)
```

```python
# Compile model
optimizer1 = Adam(learning_rate = 0.001) 
model8.compile(optimizer = optimizer1, loss = 'binary_crossentropy', metrics = ['accuracy', 'Recall'])
```

```python
history8 = model8.fit(train_data, batch_size = 256, epochs = 50, validation_data = val_data)
```

As set out under the model framework, test accuracy score was identified as the primary indicator of model performance. However, none of the models achieved the required test accuracy score of 90% based on the set benchmark. From observation, the training accuracy scores across all models were able to achieve (close to) 100% after a certain number of epochs while the corresponding test accuracy score failed to surpass the ~80%+ mark, signalling to a classic case of overfitting. Data augmentation (via Keras' ImageDataGenerator) and stopping training early (via early stopping method) did not improve the model performance, suggesting the need for an even larger training set in a bid to alleviate overfitting. Even for the VGG16 model which has the (relative) highest accuracy sore of ~80%, it is not application-ready (assuming autonomous vehicles) as there is a 1 in 5 chance of the vehicle failing to identify a pedestrian and thus, posing a serious road hazard.   

Conversely, the sensitivity (i.e. recall) scores fared much better with all models (except two) meeting the established threshold of 80%, which means that false negatives are minimized and importantly so as failing to predict an object as a person when the object is indeed a person would invariably result in accidents. In a similar vein, the F1-scores returned were largely acceptable; about half the models complied with the benchmark of 80% and for models that did not, the figures did not deviate too significantly.  

Comparing the metrics across all models, the best performing model was the VGG16 model as it not only had the highest test accuracy score and F1-score, but minimized the loss function the most as well, although the recall score may be interpreted as average. By and large, tuning/regularization did not improve the base 'built-from-scratch' CNN model noticeably. While some models performed fairly well for a handful of metrics, said models were ultimately let down by the other metrics. Thus, **the VGG16 model is selected as the final production model**, with parameters summarized as follow:
