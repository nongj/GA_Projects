## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

```python
# Code: 

# Create function to calculate average
def avg(values): 
    
# Start empty counter for sum of total values

    sum_values = 0

# Create loop to sum the individual values

    for value in values:
        
        sum_values += value

# Calculate the average

        average = sum_values / len(values)
        
    return(average)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code: 

# Create function to calculate SD
def stdd(values):
        
# Start empty counter for sum of deviations and sum of deviations over the population size

    sum_dist = 0
    
    sum_dist_over_n = 0

# Create loop to calculate the total sum of deviations over the population size

    for value in values:
        
        dist_sq = (value - avg(values)) ** 2
        
        sum_dist += dist_sq
        
        sum_dist_over_n = sum_dist * (1 / len(values))
        
# Calculate SD using sqroot of the above result
        
        std_dev = sum_dist_over_n ** (1 / 2)
        
    return(std_dev)
```

```python
stdd(a)
```

## Dictionary Comprehension

```python
# Dictionary Comprehension (Applied to 'total_2017', 'total_2018', 'total_2019', 'total_avg')
sd_total_2017 = {'sd_total_2017':stdd(main_df['total_2017']) for value in main_df['total_2017']}

sd_total_2018 = {'sd_total_2018':stdd(main_df['total_2018']) for value in main_df['total_2018']}

sd_total_2019 = {'sd_total_2019':stdd(main_df['total_2019']) for value in main_df['total_2019']}

sd_total_avg = {'sd_total_avg':stdd(main_df['total_avg']) for value in main_df['total_avg']}

print(sd_total_2017)

print(sd_total_2018)

print(sd_total_2019)

print(sd_total_avg)
```

```python
# Comparing with the SDs for the SAT scores in the summary statistics, there are minor differences in the SD values as the .describe() function uses the pd.std() function to calculate SD. 
 # By default, the pd.std() function adopts 'N - 1' instead of 'N' as the population size. The SDs are normalized by N-1 to better estimate the variance of a population. 
    # By using 'len(values) - 1' (for consistency with pd's std.() function) instead of 'len(values)' solely, the computed SD equals to the SD values calculated via pd.std(). 

def stdd1(values):
        
    sum_dist = 0
    
    sum_dist_over_n = 0

    for value in values:
        
        dist_sq = (value - avg(values)) ** 2
        
        sum_dist += dist_sq
        
        sum_dist_over_n = sum_dist * (1 / (len(values) - 1))
        
        std_dev = sum_dist_over_n ** (1 / 2)
        
    return(std_dev)

# By applying the revised SD function to 2017 total average SAT scores as an example, the returned SD value (89.91 marks) corresponds with the respective SD value in the summary statistics table. 
stdd1(main_df['total_2017'])
```

```python
# Calculate coefficients of variations to measure extent of deviation in percentage
cv_total_2017 = sd_total_2017['sd_total_2017'] / main_df['total_2017'].mean()

cv_total_2018 = sd_total_2018['sd_total_2018'] / main_df['total_2018'].mean()

cv_total_2019 = sd_total_2019['sd_total_2019'] / main_df['total_2019'].mean()

cv_total_avg = sd_total_avg['sd_total_avg'] / main_df['total_avg'].mean()

print(cv_total_2017)

print(cv_total_2018)

print(cv_total_2019)

print(cv_total_avg)
```

The standard deviations for total SAT scores are relatively similar within a range of 89-96 marks. Additionally, the coefficient of variation (CV) values are approximately 8%, which translates to the respective state total scores generally not deviating more than 8% from the respective mean scores for the given year and on average. The definition of an "acceptable" CV is contingent on the industry/field. In the context of analyzing deviations in SAT scores at the state scale, CVs of ~8% are considered reasonable.
